## Preprocessing Pipeline

```python
import nltk 
nltk.download('stopwords') 
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def remove_stop(sentence):
    "Remove all the stop words from the document"
    eng_stop_words = set(stopwords.words('english')) 
    new_sentence = [word for word in sentence.split() if word not in eng_stop_words]
    return " ".join(new_sentence)
```

```python
import re

def clean_speacial(sentence):
    "remove all the unwanted characters from the sentence"
    cleaned = re.sub(r'[^A-Za-z0-9\s\.]', '', sentence)     # keep only allowed characters
    return   cleaned.strip()      
```

```python
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer


def lemmatize(sentence):
    "Porter stemmer alogirthm in pre processing"
    wnl = WordNetLemmatizer()
    new_sentence = [wnl.lemmatize(word) for word in sentence]
    return " ".join(new_sentence)
```

```python
def pre_processing_half(headlines):

    "pre processing for the Summarization tasks (till the cleaning speacialezed tokens)"
    #lower case
    print("*")
    headlines  = headlines.apply(lambda headline : headline.lower())
    print("*")
    # remove stop words
    proc_headline = headlines.apply(remove_stop)
    print("*")
    # clean speacial tokens
    proc_headline = proc_headline.apply(clean_speacial)
    print("*")
    #tokenization
    proc_headline = proc_headline.apply(word_tokenize)
    print("*")
    #lemmatization
    proc_headline = proc_headline.apply(lemmatize)
    
    

    return proc_headline
```

## Contextual Embeddings

```python
from transformers import BertTokenizer,BertModel
import torch
```

```python
def get_bert_embeddings(text_list):
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
    model = BertModel.from_pretrained("bert-base-uncased")
    inputs = tokenizer(text_list, return_tensors="pt", padding=True, truncation=True, max_length=128)
    with torch.no_grad():
        # Pass through BERT model
        outputs = model(**inputs)
    # Mean pooling across tokens to get a single vector per headline
    embeddings = outputs.last_hidden_state.mean(dim=1)
    return embeddings.numpy()
```

```python
bert_embeddings = get_bert_embeddings(pre_proc_headline.to_list()) 
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

def get_scores(y_true, y_pred):
    scores = f"""
    accuracy:, {accuracy_score(y_true,y_pred)} | precision: {precision_score(y_true, y_pred, average = 'macro')} 
    | recall: {recall_score(y_true, y_pred, average = 'macro')} | f1: {f1_score(y_true, y_pred, average = 'macro')} 
    """
    print(scores)
    return accuracy_score(y_true,y_pred)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier

scores = {}

def train_models(topic, embeddings, y_labels = y_labels, split_ratio = 0.2):
    print("############",topic,"###############")
    scores[topic] = []
    knn_model = KNeighborsClassifier(n_neighbors=5)  
    xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')

    for col, val in y_labels.items():
        print("Label : _________",col,"_________")
        X_train, X_test, y_train, y_test = train_test_split(embeddings, val, test_size=split_ratio, random_state=42)
        # KNN Model training and predicitons
        knn_model.fit(X_train, y_train)
        knn_predictions = knn_model.predict(X_test)
        print("KNN Predictions")
        scores[topic].append(get_scores(y_test, knn_predictions))
        #XGBoost Model training and predicitons
        xgb_model.fit(X_train, y_train)
        xgb_predictions = xgb_model.predict(X_test)
        print("XGBoost Predictions")
        get_scores(y_test, xgb_predictions)    
```

```python
import matplotlib.pyplot as plt
import numpy as np


def plot_accuracies(model1_acc, model2_acc = scores['Contextual_Embeddings']):
    # Custom x labels
    x_labels = ['Sensational', 'Victim Reference', 'Villain Reference', 'Villain', 'Victim', 'Effect Frames', 'Source Frame', 'Context Frame']

    # x values for positioning
    x = np.arange(len(x_labels))

    # Width of the bars
    width = 0.35

    # Plot bars
    fig, ax = plt.subplots(figsize=(10, 6))  # Adjust the figure size for better readability
    bar1 = ax.bar(x - width/2, model1_acc, width, label='Model')
    bar2 = ax.bar(x + width/2, model2_acc, width, label='Baseline')

    # Labeling
    ax.set_xlabel('Categories')
    ax.set_ylabel('Accuracy')
    ax.set_title('Comparison of model and Baseline')
    ax.set_xticks(x)
    ax.set_xticklabels(x_labels, rotation=45, ha='right')  # Rotate labels for readability
    ax.legend()

    # Display the plot
    plt.tight_layout()  # Adjust layout for readability
    plt.show()
```

```python
from transformers import pipeline

# Load a pre-trained NER pipeline
ner_pipeline = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english", aggregation_strategy="simple")


def ner(text = ''):
    entities = ner_pipeline(text)
    return 1 if len(entities) > 0 else 0
```

### Sentiment Scores

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def calc_sentiment(text = ''):
    sid_obj = SentimentIntensityAnalyzer()
 
    # polarity_scores method of SentimentIntensityAnalyzer
    # object gives a sentiment dictionary.
    # which contains pos, neg, neu, and compound scores.
    sentiment_dict = sid_obj.polarity_scores(text)
 
    # decide sentiment as positive, negative and neutral
    if sentiment_dict['compound'] >= 0.05 :
        return 1
 
    elif sentiment_dict['compound'] <= - 0.05 :
        return -1
 
    else :
        return 0
```

```python
sentiment = pre_proc_headline.apply(lambda headline : calc_sentiment(headline))
sentiment.value_counts() 
```

### Parts of Speach tags presence of adjectives

```python
import nltk
from nltk import word_tokenize, pos_tag


def adj_freq(text = ''):

    # Tokenize and POS-tag the text
    words = word_tokenize(text)
    pos_tags = pos_tag(words)

    # Extract and display adjectives
    adjectives = [word for word, pos in pos_tags if pos in ["JJ", "JJR", "JJS"]]
    return len(adjectives)
```

```python
adj = pre_proc_headline.apply(lambda headline : adj_freq(headline))
adj.value_counts() 
```

# Task 4 Finetuning

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from transformers import DataCollatorWithPadding
import torch
import evaluate
from torch.utils.data import Dataset

# Load BERT tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Define a custom dataset
class TextDataset(Dataset):
    def __init__(self, texts, labels, tokenizer):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts.iloc[idx]
        label = self.labels[idx]
        encoding = self.tokenizer(text, truncation=True, padding="max_length", max_length=128, return_tensors="pt")
        item = {key: val.squeeze() for key, val in encoding.items()}
        item["labels"] = torch.tensor(label)
        return item


# Set up Trainer with TrainingArguments
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
)

metric = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)
```

```python
def finetune(train_datset, val_dataset, num_labels, compute_metrics = compute_metrics, training_args = training_args, tokenizer = tokenizer):   
    
    model = AutoModelForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=num_labels)


    # Define data collator to handle padding
    data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

    # Initialize Trainer
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
        tokenizer=tokenizer,
        data_collator=data_collator,
        compute_metrics=compute_metrics,
    )

    # Fine-tune the model
    trainer.train()
    
    return model
```

```python
def bert_predict(texts, model):
    # Tokenize and prepare inputs
    encoding = tokenizer(texts, return_tensors="pt", padding=True, truncation=True, max_length=128)

    # Run the model in evaluation mode
    model.eval()
    with torch.no_grad():
        outputs = model(**encoding)
    
    # Get the logits (output scores before softmax)
    logits = outputs.logits
    # Apply softmax to get probabilities
    probs = torch.nn.functional.softmax(logits, dim=-1)
    # Get the predicted class (label) with the highest probability
    predictions = torch.argmax(probs, dim=-1)
    
    return predictions
```

```python
def finetune_models():
    
    bert_models = {}

    for col, value in y_labels.items():
        # Convert labels to a single integer for each row
        labels = np.argmax(value.values, axis=1)

        # Split data into training and validation sets
        train_texts, val_texts, train_labels, val_labels = train_test_split(
            pre_proc_headline, labels, test_size=0.2, random_state=42
        )

        # Prepare datasets
        train_dataset = TextDataset(train_texts, train_labels, tokenizer)
        val_dataset = TextDataset(val_texts, val_labels, tokenizer)

        # Load BERT model for classification
        num_labels = value.shape[1]  # Number of classes
        print("Label : _________",col,"_________")
        bert_models[col] = finetune(train_dataset, val_dataset, num_labels)
        
        val_predicts = bert_predict(val_texts, bert_models[col])
        scores['bert_finetune'].append(get_scores(val_predicts, val_labels))

finetune_models()
```
